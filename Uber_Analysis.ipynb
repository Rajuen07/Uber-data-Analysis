```python
# Identify popular start points - top 10
start_group=df.groupby(by="START")["MILES"].sum()
start_group.nlargest(10)
```

```python
# Identify popular stop destinations - top 10
stop_group=df.groupby(by="STOP")["MILES"].sum()
stop_group.nlargest(10)
```

```python
# start and stop pair
start_stop=df.groupby(by=["START","STOP"])["MILES"].count()
start_stop.nlargest(10)
```

```python
# The most popular start to destination pair is Morrisville-Cary
```

```python
## 2. Manipulating date & time objects
df.head()
```

```python
# Create columns by converting the start and end date into a datatime format
df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])
df.dtypes
```

```python
df.head()
```

```python
# Which month did he get most drives  ? 
df["MONTH"]=pd.DatetimeIndex(df["START_DATE"]).month
df.replace({"MONTH":{1:"jan",2:"feb",3:"march",4:"april",5:"may",6:"jun",7:"july",8:"aug",9:"sept",10:"oct",11:"nov",12:"dec"}},inplace=True)
monthly_ride=df.groupby(by="MONTH")["MONTH"].count()
monthly_ride.nlargest(1)
```

```python
df.groupby(by="MONTH")["MILES"].mean()
```

```python
# Which day did he get most drives  ? 
```

```python




df["Day"]=pd.DatetimeIndex(df["START_DATE"]).day
df
```

```python
day_rides=pd.pivot_table(df,values="MILES",index="Day",aggfunc="count").reset_index()
day_rides[day_rides.MILES==31]
```
